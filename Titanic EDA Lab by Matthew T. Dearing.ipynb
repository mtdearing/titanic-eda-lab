# EDA of the Titanic Data

```python
# Variable: PassengerId
# 
# Type = continuous
```

```python
# No missing values
titanic[titanic.PassengerId.isnull()]
```

```python
# Statistical Values
titanic.PassengerId.describe()
# Note: since this is simply an ascending set of numbers, the only meaninful value here is the
# Maximum representing the total number. If, however, these ID numbers actually represented real IDs assigned to
# actual passengers, then it would be interesting to see if any values were missing from the set.
```

```python
# Histogram
titanic.PassengerId.hist(bins=1)
# Again, since this value is a simple incrementing variable, a distribution of the data is not informative.
```
